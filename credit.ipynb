```python
# chi square test for two categorical variables i.e gender and granted
from scipy.stats import chi2_contingency
```

```python
alpha = 0.05
def p_value(p):
    if p < alpha:
        print('Reject the Null Hypothesis')
    else:
        print('Null hypothesis is not rejected')
```

```python
def chi(x):
    table = pd.crosstab(x, app_rec_new.granted)
    stat, p, dof, expected = chi2_contingency(table)
    return stat, p, dof, expected
```

```python
stat_gender, p_gender, dof_gender, expected_gender = chi(app_rec_new.CODE_GENDER)
stat_car, p_car, dof_car, expected_car = chi(app_rec_new.FLAG_OWN_CAR)
stat_realty, p_realty, dof_realty, expected_realty = chi(app_rec_new.FLAG_OWN_REALTY)
stat_income, p_income, dof_income, expected_income = chi(app_rec_new.NAME_INCOME_TYPE)
stat_edu, p_edu, dof_edu, expected_edu = chi(app_rec_new.NAME_EDUCATION_TYPE)
stat_family, p_family, dof_family, expected_family = chi(app_rec_new.NAME_FAMILY_STATUS)
stat_house, p_house, dof_house, expected_house = chi(app_rec_new.NAME_HOUSING_TYPE)
stat_mobile, p_mobile, dof_mobile, expected_mobile = chi(app_rec_new.FLAG_MOBIL)
stat_phone, p_phone, dof_phone, expected_phone = chi(app_rec_new.FLAG_WORK_PHONE)
stat_email, p_email, dof_email, expected_email = chi(app_rec_new.FLAG_EMAIL)
```

```python
cat = ['gender', 'car', 'realty', 'income', 'edu', 'family', 'house', 'mobile', 'phone', 'email']
p_list = [p_gender, p_car, p_realty, p_income, p_edu, p_family, p_house, p_mobile, p_phone, p_email]
cat = pd.DataFrame(cat, columns=['cat'])
p_list = pd.DataFrame(p_list, columns=['p'])
df_p = cat.join(p_list)
df_p
```

```python
null_p = []
for x in df_p.p:
    if x < alpha:
        null_p.append(True)
    else:
        null_p.append(False)
null_p
        
```

```python
from sklearn import preprocessing
def label(x):
    label_encoder = preprocessing.LabelEncoder()
    label_encoder.fit(x)
    return label_encoder.transform(x)
car = label(app_rec_new.FLAG_OWN_CAR)
realty = label(app_rec_new.FLAG_OWN_REALTY)
income = label(app_rec_new.NAME_INCOME_TYPE)
edu = label(app_rec_new.NAME_EDUCATION_TYPE)
family_status = label(app_rec_new.NAME_FAMILY_STATUS)
housing = label(app_rec_new.NAME_HOUSING_TYPE)
phone = label(app_rec_new.FLAG_WORK_PHONE)
email = label(app_rec_new.FLAG_EMAIL)
car
```

```python
def dataframe(c):
    return pd.DataFrame(c)

```

```python
c1 = []
index = pd.Index(range(438510))
c1 = pd.DataFrame(c1, columns=['car', 'realty', 'income_type', 'edu', 'family', 'housing', 'phone', 'email'], index=index)
c1
```

```python
c1['car'] = dataframe(car)
c1['realty'] = dataframe(realty)
c1['income_type'] = dataframe(income)
c1['edu'] = dataframe(edu)
c1['family'] = dataframe(family_status)
c1['housing'] = dataframe(housing)
c1['phone'] = dataframe(phone)
c1['email'] = dataframe(email)
c1
```

```python
c1 = c1.join(app_rec_new.granted)
c1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

y = c1.granted
features = ['car', 'realty', 'income_type', 'edu', 'family', 'housing', 'phone', 'email']
X = c1[features]

train_X, val_X, train_y, val_y = train_test_split(X,y, random_state = 0)

```

```python
from sklearn.tree import DecisionTreeClassifier

model_d = DecisionTreeClassifier()
model_d.fit(train_X, train_y)
pred_d = model_d.predict(val_X)

pred_d = pd.DataFrame(pred_d, columns=['pred_d'])
pred_d_1 = pred_d.loc[pred_d['pred_d'] == 1]
pred_d_1
```

```python
score = model_d.score(val_X, val_y)
score
```

```python
from sklearn.ensemble import RandomForestClassifier

model_r = RandomForestClassifier()
model_r.fit(train_X, train_y)
pred_r = model_r.predict(val_X)

pred_r = pd.DataFrame(pred_r, columns=['pred'])
pred_r
```
